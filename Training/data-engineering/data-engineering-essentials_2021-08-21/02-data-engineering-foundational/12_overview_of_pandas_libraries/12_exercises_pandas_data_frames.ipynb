```python
#exercise 4 Get all the records from orders where there are no corresponding records in order_items
orders.set_index('order_id'). \
join(order_items.set_index('order_item_order_id')). \
query('order_item_id.isna()')
```

```python
#exercise 5 Get all the customers who have not placed any orders
customers.set_index('customer_id'). \
join(orders.set_index('order_customer_id'),how='left'). \
query('order_id.isna()')
```

```python
#exercise 6 Get the revenue by status
orders.set_index('order_id'). \
join(order_items.set_index('order_item_order_id'),how='inner'). \
groupby('order_status')['order_item_subtotal'].agg(['sum']).round(2).rename(columns={'sum':'revenue'})
```
