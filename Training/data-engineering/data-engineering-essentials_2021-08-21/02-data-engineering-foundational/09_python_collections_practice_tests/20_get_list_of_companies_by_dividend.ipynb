### Step 2: Provide the solution

Now come up with the solution by developing the required logic. Once the function is developed, go to the next step to take care of the validation.

````python
def get_companies_by_dividend_payout(companies):
    # Your code should go here
    divide_companies={}
    for company in companies:
        if company['Dividend']>0:  # creating the dictionary by dividing dividend paying and non_dividend paying
            if 'Dividend Paying' in divide_companies.keys():
                divide_companies['Dividend Paying'].append(company)  
            else:
                divide_companies['Dividend Paying']=[company]
        else:
            if 'Non Dividend Paying' in divide_companies.keys():
                divide_companies['Non Dividend Paying'].append(company)
            else:
                divide_companies['Non Dividend Paying']=[company]
    #creating two list by divided data
    dividend_companies=divide_companies['Dividend Paying'] 
    non_dividend_companies=divide_companies['Non Dividend Paying']
    companies_by_dividend_payout={}
    #performing sorting on dividend and non_dividend
    companies_by_dividend_payout['Non Dividend Paying']=sorted(non_dividend_companies,key=lambda company:company['Price'])
    companies_by_dividend_payout['Dividend Paying']=sorted(dividend_companies,key=lambda company:(company['Dividend'],company['Price']))
    return companies_by_dividend_payout
````
